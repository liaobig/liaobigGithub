```python
from tkinter import *
from tkinter import ttk
from enum import Enum
class State(Enum):
    calculated=0
    in_number1=1
    post_operator=2
    in_number2=3
    
class Calculator(Tk):
    def __init__(self,*args,**kwargs):
        super().__init__(*args,**kwargs)
        self.title("Simple calculator")
        
        self.result_string=StringVar()
        self.result_string.set("0")
        self.number1="0"
        self.number2=None
        self.operator=None
        self.state=State.calculated
        
        output=ttk.Label(self,textvariable=self.result_string,font=("TKDefaultFont",48),anchor=E)#E：东边对齐，右对齐
        output.pack(side=TOP,expand=True,fill=BOTH)  
        
        row = Frame(self)
        row.pack(side=TOP,expand=True, fill=BOTH)
        
        self.add_button(row,"CE",self.on_ce_click)
        self.add_button(row,"C",self.on_c_click)
        self.add_button(row,"BACK",self.on_back_click)
        self.add_button(row,"/",self.on_chu_click)
        #for i in range(4):
            #ttk.Button(row,text=i).pack(side=LEFT,expand=True,fill=BOTH)
        row = Frame(self)
        row.pack(side=TOP,expand=True, fill=BOTH)
        self.add_button(row,"7",self.on_7_click)
        self.add_button(row,"8",self.on_8_click)
        self.add_button(row,"9",self.on_9_click)
        self.add_button(row,"*",self.on_cheng_click)
        
        row = Frame(self)
        row.pack(side=TOP,expand=True, fill=BOTH)
        
        #self.add_button(row,"4",self.on_4_click)
        self.add_button(row,"4",self.gen_on_number_click("4"))
        
        self.add_button(row,"5",self.on_5_click)
        self.add_button(row,"6",self.on_6_click)
        self.add_button(row,"-",self.on_jian_click)
        
        row = Frame(self)
        row.pack(side=TOP,expand=True, fill=BOTH)
        self.add_button(row,"1",self.on_1_click)
        self.add_button(row,"2",self.on_2_click)
        self.add_button(row,"3",self.on_3_click)
        self.add_button(row,"+",self.on_jia_click)
        
        row = Frame(self)
        row.pack(side=TOP,expand=True, fill=BOTH)
        self.add_button(row,"+/-",self.on_zf_click)
        self.add_button(row,"0",self.on_0_click)
        self.add_button(row,".",self.on_dian_click)
        self.add_button(row,"=",self.on_deng_click)
        
    def add_button(self,parent,text=None,command=None):
        button=ttk.Button(parent,text=text,command=command)
        button.pack(side=LEFT,expand=True,fill=BOTH)
    
    def on_ce_click(self):
        print("ce")
    def on_c_click(self):
        print("c")
    def on_back_click(self):
        print("back")
    
    def gen_on_number_click(self,ch):
        return lambda:self.on_number(ch)

    def on_7_click(self):
        self.on_number("7")
        print("7")
    def on_8_click(self):
        self.on_number("8")
        print("8")
    def on_9_click(self):
        self.on_number("9")
        print("9")
    def on_5_click(self):
        self.on_number("5")
        print("5")
    def on_6_click(self):
        self.on_number("6")
        print("6")
    def on_4_click(self):
        self.on_number("4")
        print("4")
    def on_3_click(self):
        self.on_number("3")
        print("3")
    def on_2_click(self):
        self.on_number("2")
        print("2")
    def on_1_click(self):
        self.on_number("1")
        print("1")
    def on_0_click(self):
        self.on_number("0")
        print("0")
    def on_number(self,ch):
        if self.state == State.calculated:
            self.number1=ch
            self.state=State.in_number1
            self.result_string.set(self.number1)
        elif self.state==State.in_number1:
            self.number1=self.number1+ch
        elif self.state==State.post_operator:
            self.number2=ch
            self.state=State.in_number2
            self.result_string.set(self.number2)
        elif self.state==State.in_number2:
            self.number2=self.number2+ch
        else:assert(0)
            
    def on_operator(self,operator):
        if self.state == State.calculated:   
            self.state=State.post_operator  
        elif self.state==State.in_number1:    
            self.state=State.post_operator      
        #elif self.state==State.post_operator:     
        elif self.state==State.in_number2:
            self.on_deng_click()
            self.state=State.post_operator
        else:assert(0)
        self.operator=operator    
        
    def on_chu_click(self):
        self.on_operator("/")
        print("/")    
    def on_cheng_click(self):
        self.on_operator("*")
        print("*")
    def on_jian_click(self):
        self.on_operator("-")
        print("-")
    def on_jia_click(self):
        self.on_operator("+")
        print("+")
    def on_deng_click(self):
        if self.state!=State.in_number2:
            return
        print("=",end="")
        if self.operator=="+":
            self.number1=str(float(self.number1)+float(self.number2))
        if self.operator=="-":
            self.number1=str(float(self.number1)-float(self.number2))
        if self.operator=="*":
            self.number1=str(float(self.number1)*float(self.number2))
        if self.operator=="/":
            self.number1=str(float(self.number1)/float(self.number2))
            
        self.result_string.set(self.number1)
        self.state=State.calculated
        
    def on_zf_click(self):
        print("+/-")
    def on_dian_click(self):
        self.on_number(".")
        print(".")
        
if __name__ == "__main__":
    app=Calculator()
    app.mainloop()
```
